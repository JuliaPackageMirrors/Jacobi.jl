```julia
#fun(x,k=1) = cos(k*x) + 1
#ifun(x,k=1) = x + sin(k*x) / k
fun(x) = cos(x)
ifun(x) = -x.*sin(x) + sin(x) - cos(x)
```

```julia
function calc_integr{T<:Number}(Q, ::Type{T}=Float64)
    z = Jacobi.zglj(Q, 1, 0, T)
    w = Jacobi.wglj(z, 1, 0,)
    
    y = fun(z)
    
    return sum(w .* y)
end

calc_integr(50, BigFloat)
```

```julia
set_bigfloat_precision(256)

Iex{T<:Number}(::Type{T}=Float64) = ifun(one(T)) - ifun(-one(T))
Iex(BigFloat)
```

```julia
Q = 2:40
tipo =BigFloat
Ie = Iex(tipo)
err = [abs(calc_integr(q, tipo) - Ie) for q in Q];
```
